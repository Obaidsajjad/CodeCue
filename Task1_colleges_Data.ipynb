```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("All the Intermediate Colleges in Pakistan.csv")
```

```python
df.head()
```

```python
# df["Location"].replace("N/A",np.nan)
df["Location"].dropna(inplace=True)
df.isnull().sum()
```

```python
import numpy as np
```

```python
df.drop("Affiliation",axis=1, inplace=True)
df.head()
```

```python
dummy=pd.get_dummies(df["Sector"])
df=pd.concat([df,dummy],axis=1)
df.head()
```

```python
df["Sector"].fillna()
```

```python
df=df.replace({"Public":1, "Private":0})
df.head()
```

```python
# @title Number of colleges by sector

df.groupby('Sector').size().plot(kind='pie')
```

```python
# @title Private vs Public Colleges

df.groupby(['Private', 'Public']).size().unstack().plot(kind='bar')
```

```python
# @title Sector

from matplotlib import pyplot as plt
df['Sector'].plot(kind='hist', bins=20, title='Sector')
plt.gca().spines[['top', 'right',]].set_visible(True)
```

```python
df["Sector"].fillna(df["Sector"].median(),inplace=True)
```
