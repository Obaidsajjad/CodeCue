```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Suicide_bombing.csv')
df.head(15)
```

```python
df.shape
```

```python
df.isnull().sum()*100/df.shape[0]
```

```python
df.drop(['Holiday Type','Influencing Event/Event','Explosive Weight (max)','Hospital Names'],axis=1,inplace=True)
df.head(15)
```

```python
df.isnull().sum()
# df.shape
```

```python
df.drop(columns=['Targeted Sect if any'],axis=1, inplace=True)
df.drop(columns=['Time'],axis=1, inplace=True)
df.drop(columns=['Islamic Date'],axis=1, inplace=True)
```

```python
df["Injured Max"] = pd.to_numeric(df["Injured Max"], errors="coerce")
```

```python
df["Killed Min"].fillna(df["Killed Min"].mean(), inplace=True)
df["Injured Max"].fillna(df["Injured Max"].mean(), inplace=True)
df["Injured Min"].fillna(df["Injured Min"].mean(), inplace=True)
```

```python
df["No. of Suicide Blasts"].fillna(df["No. of Suicide Blasts"].median(), inplace=True)
df["Temperature(C)"].fillna(df["Temperature(C)"].mean(), inplace=True)
df["Temperature(F)"].fillna(df["Temperature(F)"].mean(), inplace=True)
```

```python
open=pd.get_dummies(df['Open/Closed Space'],drop_first=True)
blaseDaytype=pd.get_dummies(df['Blast Day Type'],drop_first=True)
```

```python
df=pd.concat([df,open,blaseDaytype],axis=1)
# df.drop(['Open/Closed Space','Blast Day Type'],axis=1,inplace=True)
df.head(15)
```
